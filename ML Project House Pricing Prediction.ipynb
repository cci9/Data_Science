```python

    Q1 = np.percentile(dataset['RM'], 25,
                   interpolation = 'midpoint')
 
    Q3 = np.percentile(dataset['RM'], 75,
                       interpolation = 'midpoint')
    IQR = Q3 - Q1
    # Upper bound
    upper=Q3+1.5*IQR
    upper_array=np.array(dataset['RM']>=upper)
    # Lower bound
    lower=Q1-1.5*IQR
    lower_array=np.array(dataset['RM']<=lower)

    dataset = dataset[(dataset['RM'] <= upper) & (dataset['RM'] >= lower)]
```

```python

    Q1 = np.percentile(dataset['PTRATIO'], 25,
                   interpolation = 'midpoint')
 
    Q3 = np.percentile(dataset['PTRATIO'], 75,
                       interpolation = 'midpoint')
    IQR = Q3 - Q1
    # Upper bound
    upper=Q3+1.5*IQR
    upper_array=np.array(dataset['PTRATIO']>=upper)
    # Lower bound
    lower=Q1-1.5*IQR
    lower_array=np.array(dataset['PTRATIO']<=lower)

    dataset = dataset[(dataset['PTRATIO'] <= upper) & (dataset['PTRATIO'] >= lower)]
```

```python

    Q1 = np.percentile(dataset['LSTAT'], 25,
                   interpolation = 'midpoint')
 
    Q3 = np.percentile(dataset['LSTAT'], 75,
                       interpolation = 'midpoint')
    IQR = Q3 - Q1
    # Upper bound
    upper=Q3+1.5*IQR
    upper_array=np.array(dataset['LSTAT']>=upper)
    # Lower bound
    lower=Q1-1.5*IQR
    lower_array=np.array(dataset['LSTAT']<=lower)

    dataset = dataset[(dataset['LSTAT'] <= upper) & (dataset['LSTAT'] >= lower)]
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

# R square and Adjusted R Square
R^2 = 1 -SSR/SST
Adjusted R^2 = 1-(1-R^2)(n-1)/(n-k-1)

where R^2 is the coefficient of determination
and Adjusted R^2 is always less than R^2
n: no of observation
k: no. of predictor variables

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
adjusted_score = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
adjusted_score
```
